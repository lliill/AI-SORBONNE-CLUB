```python
#Euler schema function:
def F(state, control) -> "R^6 -> R^4":
    x, y, theta, v = adnumber(state)
    w, a = adnumber(control)
    f = h*v
    b = f*np.cos(w) + d - np.sqrt(d**2 - f**2 * np.sin(w)**2)

    x = x + b*np.cos(theta)
    y = y + b*np.sin(theta)
    theta = theta + np.arcsin(np.sin(w) * f/d)
    v = v + h*a
    
    return np.array([x, y, theta, v])
```
